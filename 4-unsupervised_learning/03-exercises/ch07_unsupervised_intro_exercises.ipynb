## Setup: Load the Digits Dataset

```python
digits = load_digits()
X = digits.data    # 64 features per image
y = digits.target  # true digit label (0–9) — HIDE this during exploration!

print(f'Shape: {X.shape} → {X.shape[0]} digit images, {X.shape[1]} pixel features')
print('Classes: digits 0–9')

# Show a few example images
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, (ax, img, label) in enumerate(zip(axes.flat, X[:10], y[:10])):
    ax.imshow(img.reshape(8, 8), cmap='gray_r')
    ax.set_title(f'Label: {label}', fontsize=12)
    ax.axis('off')
plt.suptitle('Sample digit images (64-dimensional data)', fontsize=13)
plt.tight_layout()
plt.show()
```
