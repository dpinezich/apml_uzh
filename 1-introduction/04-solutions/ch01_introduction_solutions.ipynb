```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
np.random.seed(42)

df = sns.load_dataset('penguins')
print('Dataset loaded! Shape:', df.shape)
```

## Task 1 Solution: Inspect the Dataset

```python
# 1a: Shape
print('Shape:', df.shape)
print(f'→ {df.shape[0]} penguins, {df.shape[1]} columns')
```

```python
# 1b: Data types and non-null counts
df.info()
```

```python
# 1c: Missing values
print('Missing values per column:')
print(df.isnull().sum())
print(f'\nComplete rows: {df.dropna().shape[0]} of {df.shape[0]}')
print('→ 11 rows have at least one missing value.')
```

```python
# 1d: Species counts
print('Penguins per species:')
print(df['species'].value_counts())
print('\n→ Adelie is the most common species (152 penguins)')
```

## Task 2 Solution: Summary Statistics

```python
df.describe().round(2)
```

## Task 3 Solution: Visualize Feature Distributions

```python
# 3a: Histograms
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

for ax, feature in zip(axes.flat, features):
    sns.histplot(data=df, x=feature, hue='species', ax=ax, kde=True, alpha=0.5)
    ax.set_title(f'Distribution of {feature}', fontsize=12)

plt.suptitle('Feature Distributions by Penguin Species', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print('Observation: Gentoo penguins tend to have longer flippers and larger body mass.')
print('Adelie and Chinstrap overlap more in bill length but differ in bill depth.')
```

```python
# 3b: Scatter plot
fig, ax = plt.subplots(figsize=(9, 6))
colors = {'Adelie': '#e74c3c', 'Chinstrap': '#3498db', 'Gentoo': '#2ecc71'}

for species, group in df.groupby('species'):
    ax.scatter(group['bill_length_mm'], group['flipper_length_mm'],
               label=species, color=colors[species], alpha=0.7, s=60, edgecolors='white')

ax.set_xlabel('Bill Length (mm)', fontsize=12)
ax.set_ylabel('Flipper Length (mm)', fontsize=12)
ax.set_title('Bill Length vs Flipper Length by Species', fontsize=13)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

print('\nGentoo is clearly separable (top-right cluster).')
print('Adelie and Chinstrap overlap more — a model will need to use all features.')
```

## Task 4 Solution: Connecting to the DS Workflow

```python
# Mean bill length per species
mean_bill = df.groupby('species')['bill_length_mm'].mean().round(2)
print('Mean bill length (mm) per species:')
print(mean_bill)
print(f'\n→ Chinstrap has the longest bill on average ({mean_bill.max():.1f} mm)')
```

## Bonus Solution: Pair Plot

```python
pair_plot = sns.pairplot(
    df.dropna(),
    hue='species',
    palette={'Adelie': '#e74c3c', 'Chinstrap': '#3498db', 'Gentoo': '#2ecc71'},
    diag_kind='kde',
    plot_kws={'alpha': 0.6, 's': 40}
)
pair_plot.fig.suptitle('Pair Plot — All Feature Relationships (Penguins)', y=1.02, fontsize=13)
plt.show()

print('Best separating pair: bill_length_mm vs bill_depth_mm or')
print('                      flipper_length_mm vs body_mass_g')
print('Gentoo is always clearly separated; Adelie/Chinstrap are harder to distinguish.')
```
